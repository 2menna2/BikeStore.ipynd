##### Is there a correlation between total_units sold and revenue?

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Scatter plot with regression line
plt.figure(figsize=(10, 6))
sns.regplot(data=df, x='total_units', y='revenue', scatter_kws={'alpha': 0.6}, line_kws={'color': 'red'})

# Adding titles and labels
plt.title('Correlation between Total Units Sold and Revenue', fontsize=16)
plt.xlabel('Total Units Sold', fontsize=14)
plt.ylabel('Revenue', fontsize=14)
plt.grid(alpha=0.3)
plt.show()
```

##### What are the busiest months or days based on the number of orders?

```python
# Convert order_date to datetime format
df['order_date'] = pd.to_datetime(df['order_date'])

# Extract month and day of the week
df['month'] = df['order_date'].dt.month
df['day_of_week'] = df['order_date'].dt.day_name()  # Gives the day name like 'Monday', 'Tuesday', etc.
```

```python
# Group by month
busiest_months = df.groupby('month')['order_id'].count().sort_values(ascending=False)

# Group by day of the week
busiest_days = df.groupby('day_of_week')['order_id'].count().sort_values(ascending=False)

print("Orders by Month:")
print(busiest_months)

print("\nOrders by Day of the Week:")
print(busiest_days)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Bar plot for months
plt.figure(figsize=(10, 6))
sns.barplot(x=busiest_months.index, y=busiest_months.values, palette="Blues_d")
plt.title('Busiest Months Based on Number of Orders', fontsize=16)
plt.xlabel('Month', fontsize=14)
plt.ylabel('Number of Orders', fontsize=14)
plt.show()
```

##### How does performance vary between stores in different state or city?

```python
# Group by store and state, and calculate total revenue or units sold
store_performance = df.groupby(['store_name', 'state'])['revenue'].sum().reset_index()

# Alternatively, group by store and city
store_performance_city = df.groupby(['store_name', 'city'])['revenue'].sum().reset_index()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Bar plot for store vs state (Revenue)
plt.figure(figsize=(12, 8))
sns.barplot(data=store_performance, x='store_name', y='revenue', hue='state', palette='viridis')
plt.title('Store Performance Across Different States Based on Revenue', fontsize=16)
plt.xlabel('Store Name', fontsize=14)
plt.ylabel('Total Revenue', fontsize=14)
plt.xticks(rotation=45)
plt.legend(title='State', loc='upper left')
plt.show()
```

##### Who are the top 5 customers by revenue ?

```python
# Group by customer and sum the revenue
top_customers_revenue = df.groupby('customers')['revenue'].sum().sort_values(ascending=False).head(5)

# Display the result
print(top_customers_revenue)


```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Bar plot for top customers by revenue
plt.figure(figsize=(10, 6))
sns.barplot(x=top_customers_revenue.index, y=top_customers_revenue.values, palette='Blues_d')
plt.title('Top 5 Customers by Revenue', fontsize=16)
plt.xlabel('Customer', fontsize=14)
plt.ylabel('Total Revenue', fontsize=14)
plt.xticks(rotation=45)
plt.show()
```
